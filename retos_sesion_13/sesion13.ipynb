```python
# Ejemplo 13, sumar los n√∫meros mientras no se ingrese por teclado el n√∫mero 0
print ("Ejemplo 13")
suma = 0
numero = int(input("Ingrese un n√∫mero: "))
while numero != 0:
    suma += numero
    numero = int(input("Ingrese un n√∫mero: "))
print(suma)
```

```python
# Ejercicio 3, Ingresa un n√∫mero por teclado y genera un contador hasta 0,
# si el n√∫mero es negativo no hace nada
numero = int(input("Ingrese un n√∫mero: "))
while numero >= 0:
    print(numero)
    numero-=1

print("Terminado")
```

*La sentencia ***break*** es una palabra reservada, para salir de un ciclo*

```python
# Ejemplo 14, De la siguiente lista de frutas imprimir los elementos
# hasta que se encuentre un gusano üêõ con for
frutas = ['üçé','üçå','üçá','üçâ','üçä','üêõ','üçã','üçç']
print ("Ejemplo 14")
for fruta in frutas:
    if fruta == 'üêõ':
        break
    print(fruta)
print ("Fin")
```

```python
# Ejemplo 16, Crear un ciclo infinito que pida una cadena de texto
# la ponga en may√∫sculas y la imprima hasta que se ingrese la palabra salir

print("Ejemplo 16")
while True:
    texto = input("Ingrese un texto: ")
    if texto == 'salir':
        break
    print(texto.upper())
print ("Fin")
```

```python
# Ejercicio 4, Crear un ciclo infinito que reciba un n√∫mero por teclado
# y verifique si es par o impar hasta que se ingrese el n√∫mero 0

while True:
    nro = int(input("Ingrese un numero: "))
    if nro == 0:
        break
    else:
        if nro%2 == 0:
            print(f"{nro} es par")
        else:
            print(f"{nro} es impar")
print ("Fin")
```

*Estructuras por comprension*
 ***son una forma de crear estructuras en una sola linea***

```python
# Ejemplo 17, Crear una lista de los n√∫meros pares del 2 al 10
print ("Ejemplo 17")
pares = [i for i in range(2, 11, 2)]
print(pares)
```

```python
# creando un conjunto
pares = {i for i in range(2, 11, 2)}
print(pares)
print(type(pares))
```

```python
# Ejemplo 18, Crear una lista de los n√∫meros pares del 2 al 10 con condicional
print ("Ejemplo 18")
pares = [i for i in range(2, 11) if i % 2 == 0]
print(pares)
```

```python
# lista de impares
impares = [i for i in range(2, 11) if i % 2 == 1]
print(impares)
```

```python
# Ejemplo 19, Crear un diccionario n√∫meros 2 al 10
# donde si es par vale "Par" y si es impar valga "Impar"
print ("Ejemplo 19")
pares = {i: "Par" if i % 2 == 0 else "Impar" for i in range(2, 11)}
print(pares)
```

```python
# Ejercicio 5, Crear una tupla de los n√∫meros impares del 1 al 10
# usando una tupla por comprensi√≥n
impares = tuple(i for i in range(1,11,2))
print(impares)
```

```python
# Ejemplo 20, Imprimir las tablas de multiplicar del 1 y 2

print ("Ejemplo 20")
for i in range(1, 3):
    print(f"Tabla del {i}")
    print("----------------")
    for j in range(1, 11):
        print(f"{i} x {j} = {i*j}")
```

```python
# Ejemplo 21, Introducir un n√∫mero por teclado y crear
# una tabla de multiplicar de ese n√∫mero del 1 al 10,
# si se ingresa 0 termina el programa

print ("Ejemplo 21")
while True:
    numero = int(input("Ingrese un n√∫mero: "))
    if numero == 0:
        break
    print(f"Tabla del {numero}")
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero*i}")
print ("Fin")
```

```python
# Ejemplo 22, Introducir un n√∫mero por teclado
# y crear una matriz nxn con la letra X

print ("Ejemplo 22")
n = int(input("Ingrese un n√∫mero: "))
matriz = [['X' for i in range(n)] for j in range(n)]
for fila in matriz:
    for columna in fila:
        print(columna, end=" ")
    print()
print (matriz)
```

```python
# Ejercicio 6, Crear una matriz ingresando un n√∫mero por teclado
# para el tama√±o de la matriz y en cada posici√≥n colocar una tupla con (i, j)

n = int(input("Ingrese un n√∫mero: "))
matriz = [[(j,i) for i in range(n)] for j in range(n)]
for fila in matriz:
    for columna in fila:
        print(columna, end=" ")
    print()

print(matriz)
```
