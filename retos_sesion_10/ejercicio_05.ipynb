```python
# METODOS DE OPERACIONES CON CONJUNTOS, retornan un new conjunto como resultado

print ("MÃ©todo union()")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print (conjunto1, conjunto2)

union = conjunto1.union(conjunto2)
print(union)
```

```python
# # contiene los elem que estan en ambos conjuntos
print ("MÃ©todo intersection()")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ'}
print (conjunto1, conjunto2)

interseccion = conjunto1.intersection(conjunto2)
print(interseccion)
```

```python
 # Los que estÃ¡n en el 1er conjunto pero no en el 2do
print ("MÃ©todo difference(), importa el orden")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ†'}
conjunto2 = {'ğŸ•','ğŸ¡','ğŸ†'}
print ("1:",conjunto1, "2:",conjunto2)

diferencia = conjunto1.difference(conjunto2)
print("conjunto1 - conjunto2:",diferencia)
diferencia = conjunto2.difference(conjunto1)
print("conjunto2 - conjunto1:",diferencia)
```

```python
 # Contiene los elementos que estÃ¡n en un conjunto
 #  o en el otro pero no en ambos  =>  complemento de la interseccion
print ("MÃ©todo symmetric_difference()")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸˆ'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸˆ'}
print (conjunto1, conjunto2)
diferencia_simetrica = conjunto1.symmetric_difference(conjunto2)
print(diferencia_simetrica)
```

```python
 # MÃ‰TODOS DE ASIGNACIÃ“N CON OPERACIONES.- Permiten realizar operaciones
 #  con conjuntos y asignar el resultado al conjunto inicial
print ("MÃ©todo intersection_update()")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ¬'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ¬'}
print (conjunto1, conjunto2)
conjunto1.intersection_update(conjunto2)
print(conjunto1)
```

```python
 # MÃ‰TODOS DE BÃšSQUEDA [54:40]
print ("MÃ©todo issubset()")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
conjunto3 = {'ğŸ”','ğŸŸ'}
print (conjunto1, conjunto2,conjunto3)
# # Â¿El conjunto1 es subconjunto del conjunto2?  => F
print(conjunto1.issubset(conjunto2))
# # Â¿El conjunto3 es subconjunto del conjunto1?  => V
# # Todos los elem de conjunto3 estan en el conjunto1
print(conjunto3.issubset(conjunto1))
```

```python
# OPERADORES CON CONJUNTOS
# Los conjuntos soportan operadores que permiten realizar operaciones

print ("Operador |=")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ'}
conjunto2 = {'ğŸ•','ğŸ¨'}
print (conjunto1, conjunto2)
```

```python
# # lo del conjunto2 pasarlo al conjunto1
## conj1 UNION conj2 y pasarlo al conj1
conjunto1 |= conjunto2
print(conjunto1)
```

```python
# # solo el pipe | significa union
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ'}
conjunto2 = {'ğŸ•','ğŸ¨'}
print(conjunto1, conjunto2)
conjunto3 = conjunto1 | conjunto2
print(conjunto3)
```

```python
# # OPERADORES PARA OPERACIONES CON CONJUNTOS:  |  &  -  ^

print ("Operador |  union")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print (conjunto1, conjunto2)
union = conjunto1 | conjunto2
print(union)
```

```python
print ("Operador &")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print (conjunto1, conjunto2)
interseccion = conjunto1 & conjunto2
print(interseccion)
```

```python
print ("Operador - ,  retorna la diferencia")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print ("1:",conjunto1, "2:",conjunto2)
diferencia = conjunto1 - conjunto2
print("1 - 2:",diferencia)
diferencia = conjunto2 - conjunto1
print("2 - 1:",diferencia)
```

```python
print ("Operador ^ , return la dif simetrica, complemento de la interseccion")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print (conjunto1, conjunto2)
diferencia_simetrica = conjunto1 ^ conjunto2
print(diferencia_simetrica)
# # los iguales  => descartarlo de la union
```

```python
# OPERADORES PARA ASIGNACIÃ“N CON OPERACIONES
# |=   &=    -=    ^=
# opera con conjuntos y ASIGNA EL RESULTADO AL conjunto inicial
print ("Operador |= UniÃ³n")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print (conjunto1, conjunto2)
conjunto1 |= conjunto2
print(conjunto1)
```

```python
print ("Operador &= IntersecciÃ³n")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print (conjunto1, conjunto2)
conjunto1 &= conjunto2
print(conjunto1)
```

```python
print ("Operador -= Diferencia")
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print ("1:",conjunto1, "2:",conjunto2)
conjunto1 -= conjunto2
print("1 - 2:",conjunto1)
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 -= conjunto1
print("2 - 1:", conjunto2)
```

```python
print ("Operador ^= Diferencia simÃ©trica") # complemento interseccion
conjunto1 = {'ğŸ”','ğŸŸ', 'ğŸ©'}
conjunto2 = {'ğŸ•','ğŸ¨','ğŸ©'}
print (conjunto1, conjunto2)
conjunto1 ^= conjunto2
print(conjunto1)
```
