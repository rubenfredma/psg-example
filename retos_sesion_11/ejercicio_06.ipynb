```python
print ("MÃ©todo items(), retorna lista-tuplas(clave, valor)")
diccionario = {'perro': 'ğŸ¶', 'gato': 'ğŸ±', 'ave': 'ğŸ¦'}
print(diccionario)
```

```python
print ("MÃ©todo keys(), retorna lista con las claves")
diccionario = {'perro': 'ğŸ¶', 'gato': 'ğŸ±', 'ave': 'ğŸ¦'}
print(diccionario)
print()
keys = diccionario.keys()
print(keys, type(keys))
print()
keys = list(keys) # Convertir a lista
print(keys, type(keys))
print(keys[0], type(keys[0]))
```

```python
print ("MÃ©todo values()")
diccionario = {'perro': 'ğŸ¶', 'gato': 'ğŸ±', 'ave': 'ğŸ¦'}
print(diccionario)
print()
values = diccionario.values()
print(values, type(values))
print()
values = list(values) # Convertir a lista
print(values, type(values))
print(values[0], type(values[0]))
```

```python
# Verifica si una clave exxiste en el diccionario
print ("FunciÃ³n in  / not in")
diccionario = {'perro': 'ğŸ¶', 'gato': 'ğŸ±', 'ave': 'ğŸ¦'}
print(diccionario)
print()
existe = 'perro' in diccionario
print(existe, type(existe))
print()
no_existe = 'pez' not in diccionario
print(no_existe, type(no_existe))
```

```python
# Recibe items del dicc y retorna un iterador
print ("FunciÃ³n iter()")
diccionario = {'perro': 'ğŸ¶', 'gato': 'ğŸ±', 'ave': 'ğŸ¦'}
print(diccionario)
```

```python
iterador = iter(diccionario.items())
print(iterador)  # objet en memoria
print(type(iterador))
print()
# da un paso y devuelve el contenido
siguiente = next(iterador)
print(siguiente, type(siguiente))
print()
siguiente = next(iterador)
print(siguiente, type(siguiente))
print()
siguiente = next(iterador)
print(siguiente, type(siguiente))
```
