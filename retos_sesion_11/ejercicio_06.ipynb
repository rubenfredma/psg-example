```python
print ("Método items(), retorna lista-tuplas(clave, valor)")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
```

```python
items = diccionario.items()
print(items, type(items))
# internamente esta como una lista de tuplas
```

```python
items = list(items) # Convertir a lista
print(items, type(items))
print(items[0], type(items[0]))
```

```python
print ("Método keys(), retorna lista con las claves")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
print()
keys = diccionario.keys()
print(keys, type(keys))
print()
keys = list(keys) # Convertir a lista
print(keys, type(keys))
print(keys[0], type(keys[0]))
```

```python
print ("Método values()")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
print()
values = diccionario.values()
print(values, type(values))
print()
values = list(values) # Convertir a lista
print(values, type(values))
print(values[0], type(values[0]))
```

```python
print ("Método update(diccionario), add elemento al dicc")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
print()
diccionario.update({'pez': '🐠', 'perro': '🐩'})
print(diccionario)
```

```python
print ("Método update(clave=valor)")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
print()
diccionario.update(pez='🐠', perro='🐩')
print(diccionario)
```

```python
print ("Método clear(), elimina sus elementos")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
diccionario.clear()
print(diccionario)
```

```python
print ("Método pop(clave)")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
print()
gato = diccionario.pop('gato')
print(gato, type(gato))
print(diccionario)
```

```python
print ("Método popitem(), elimina el ultimo(clav,val)")
diccionario = {'perro': '🐶', 'gato': '🐱'}
print(diccionario)
par = diccionario.popitem()
print(par, type(par))
print(diccionario)
# par = diccionario.popitem()
# print(par)
```

```python
print ("Asignación por referencia")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)

copia = diccionario
print(copia)

copia['ave'] = '🎉'
print(diccionario)
print(copia)
```

```python
print ("Método copy()")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
print()
copia = diccionario.copy()
print(copia)
copia['ave'] = '♻'
print(diccionario)
print(copia)
```

```python
# FUNCIONES CON DICCIONARIOS
print ("Función len()")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
longitud = len(diccionario)
print(longitud)
```

```python
# Verifica si una clave exxiste en el diccionario
print ("Función in  / not in")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
print()
existe = 'perro' in diccionario
print(existe, type(existe))
print()
no_existe = 'pez' not in diccionario
print(no_existe, type(no_existe))
```

```python
# Recibe items del dicc y retorna un iterador
print ("Función iter()")
diccionario = {'perro': '🐶', 'gato': '🐱', 'ave': '🐦'}
print(diccionario)
```

```python
iterador = iter(diccionario.items())
print(iterador)  # objet en memoria
print(type(iterador))
print()
# da un paso y devuelve el contenido
siguiente = next(iterador)
print(siguiente, type(siguiente))
print()
siguiente = next(iterador)
print(siguiente, type(siguiente))
print()
siguiente = next(iterador)
print(siguiente, type(siguiente))
```
